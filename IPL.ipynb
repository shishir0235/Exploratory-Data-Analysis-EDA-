```python
# Univariate Analysis:
# Categorical Unordered:
#Check the count for each category in the "team1" column
df["team1"].value_counts()
```

```python
#Create a countplot to visualize the count of each category in the team1 column.
sns.countplot(data=df,x="team1",width=0.6)
plt.title("IPL Teams")
plt.ylabel("Number of Teams")
plt.show()
```

```python
#Create a countplot to visualize the count of each category in the city column.
sns.countplot(data=df,x="city")
plt.title("Cities of Matches",color="green")
plt.ylabel("Number of Teams",color="red")
plt.show()
```

```python
# To see how many times and how many players won the title 'Player of Match':
df["player_of_match"].value_counts()
```

```python
# To see the 5 best performers recorded in the dataset:
df["player_of_match"].value_counts()[0:5]
```

```python
# Bar plot of these top 5 best performers:
plt.figure(figsize=(6,5))
plt.bar(list(df["player_of_match"].value_counts()[0:5].keys()),list(df["player_of_match"].value_counts()[0:5]),color="brown",width=0.6)
plt.show()
```

```python
# Match Results:
df["result"].value_counts()
```

```python
# Histogram:
plt.title("Match Results")
plt.hist(df["result"],color="purple")
plt.show()
```

```python
# Toss Winners:
df["toss_winner"].value_counts()
```

```python
# Teams won by batting first:
bat_first=df[df["win_by_runs"]!=0]
bat_first.head()
```

```python
# Most Successful Teams :
plt.figure(figsize=(5,3))
plt.hist(bat_first["win_by_runs"])
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
# Percentage Distribution through Pie Charts:
plt.figure(figsize=(7,5))
plt.pie(list(bat_first['winner'].value_counts()),labels=list(bat_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
